# AlexNet

```python
import torchvision.models as models
import os
```

#### Chargement du modèle

```python
model = models.alexnet(pretrained=True)
```

#### Structure du modèle et stockage nécessaire

```python
print("Nombre de paramètres d'AlexNet :",sum(p.numel() for p in model.parameters()))

path = os.path.expanduser("~/.cache/torch/hub/checkpoints/alexnet-owt-7be5be79.pth")
print(f"Espace de stockage occupé par AlexNet : {os.path.getsize(path) / 1024**2 :.2f} MB")
```

```python
print(f"{'Layer':30} {'Params':>12} {'Memory (MB)':>15}")
print("-" * 60)

total_mem = 0

for name, param in model.named_parameters():
    num_params = param.numel()
    mem_mb = num_params * param.element_size() / 1024**2
    total_mem += mem_mb
    print(f"{name:30} {num_params:12,d} {mem_mb:15.2f}")

print("-" * 60)
print(f"{'TOTAL':30} {'':12} {total_mem:15.2f} MB")
```
